```python
P = (10, 11)

match P:
    case (0, 0):
        print("blue origin")
    case (1, 1):
        print("red")
    case (1, 0):
        print("orange")
    case (0, 1):
        print("dark green")
    case (x, 0):
        print(f"grey for x = {x}")
    case (0, y):
        print(f"light green for y = {y}")
    case (x, y) if x == y:
        print(f"purple for ({x}, {y})")
    case (x, y):
        print(f"yellow ({x}, {y})")
```

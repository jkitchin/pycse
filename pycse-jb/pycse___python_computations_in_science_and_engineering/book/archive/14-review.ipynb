```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.random.random(12)
y = np.random.random(12)

p = np.polyfit(x, y, 1)

with plt.xkcd():
    plt.plot(x, y, 'bo')
    plt.plot(x, np.polyval(p, x))
    plt.xlabel('x values')
    plt.ylabel('y')
```
